## 2.import

```
from selenium import webdriver
from bs4 import BeautifulSoup
```

## 3. selenium 옵션주기

```
options = webdriver.ChromeOptions()
options.add_argument('--headless') # 인터페이스 없는
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

## 4. 크롬 드라이버 실행

```
driver = webdriver.Chrome('chromedriver', options = options)

# 3초간 기다리자.
driver.implicitly_wait(3)


# 페이지에서 소스 가져오기
driver.get('https://www.zeropay.or.kr/main.do?pgmId=PGM1580')

```

# 5. 소스를 분석 가능하게 만들기

```

# 페이지 소스를 html이라는 변수에 저장하기
html = driver.page_source

# html이라는 소스에 저장한 소스를 분석가능하게 만들기
# BeautifulSoup은 문단 Html의 구성요소 분석하여 나눠주기 
soup = BeautifulSoup(html, 'html.parser') # parser 는 옵션 종류


```

```
# 잘 가져왔는지 예쁘게 출력해보기
print(soup.prettify())
```

```
# 원하는 부분만 가져오기
# 개발자도구 F12를 열어서 해당 원하는 부분을 우클릭 후 select an element 선택하여 원하는 상세 소스를 찾는다
s1 = soup.find_all('div', {'class':'question faqToggle'})
s2 = soup.find_all('div', {'class':'answer'})

```

```
# question에 해당하는 html 확인하기
print(s1)
```

```
# Answer에 해당하는 html 확인하기
print(s2)
```

```
print(len(s1))
print(len(s2))
```

```
# 가져온 것을 저장하기
#QnA={que.text.strip():ans.text.strip() for que in s1 for ans in s2}
Questions = []
Answers = []
for que in s1:
  Questions.append(que.text.strip())
  
for ans in s2:
  Answers.append(ans.text.strip())
```

```
for i in Questions:
  print(i)
  
```

```
for j in Answers:
   print(j)
```

```
# 딕셔너리로 묶기 
# QnA_dict = dict(zip(Questions, Answers))
QnA_dict = {'질문':Questions, '답변':Answers}
print(QnA_dict)
```

##6. 파일로 저장하기

```
import pandas as pd
df = pd.DataFrame(QnA_dict)
df
```
